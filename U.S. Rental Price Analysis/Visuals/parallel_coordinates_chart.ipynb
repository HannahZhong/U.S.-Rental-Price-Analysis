```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
```

```python
# Read the data from the CSV file
data = pd.read_csv("merged_data.csv")

sampled_data = data.sample(frac=0.1)

attributes = ["percent_asian", "percent_rental_units_occupied", "percent_graduate_deg", "percent_commute_public_transport", "percent_commute_30_to_59", 
              "income_per_capita", "percent_not_us_citizen", "total_firms","startup_firms","percent_commute_less_30","percent_associates","zri"]

# Convert the data into a pandas DataFrame
df = pd.DataFrame(sampled_data)

# Create a dictionary to map zri values to numerical values
unique_zri = sorted(set(df['zri']))
zri_map = {zri: i for i, zri in enumerate(unique_zri)}
df['zri_id'] = df['zri'].map(zri_map)

# Create the chart
fig = px.parallel_coordinates(
    df,
    dimensions=attributes,
    color="zri_id",
    labels={"zri_id": "ZRI"},
    color_continuous_scale=px.colors.diverging.Tealrose,
    color_continuous_midpoint=1)

# Hide the color scale that is not needed in this case
fig.update_layout(coloraxis_showscale=False)

# Show the plot
fig.show()
```
