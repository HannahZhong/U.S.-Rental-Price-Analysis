```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.arima_model import ARIMA
from pmdarima.arima import auto_arima

import warnings
warnings.filterwarnings('ignore')

sns.set()

forecasting_ready_data = []

for city in famous_cities_data['City'].unique():
    city_subset = famous_cities_data[famous_cities_data['City'] == city]
    city_subset['year'] = pd.to_datetime(city_subset['year'])
    city_subset.sort_values(by='year', inplace=True)
    city_subset.set_index('year', inplace=True)
    state = city_subset['State'].values[0]
    df = city_subset    
    df.drop(['City', 'State'], axis=1, inplace=True)
    sns.lineplot(data = df)
    plt.xticks(rotation=45)
    plt.title(str(city) + " " + 'City Rent over Time', fontsize=12)
    plt.show()
    
    

    # Split the data into training and testing sets

    # Find the optimal p, d, and q values for the ARIMA model
    model = auto_arima(df, start_p=0, start_q=0,
                       max_p=5, max_q=5, m=12,
                       start_P=0, seasonal=True,
                       d=None, D=1, trace=True,
                       error_action='ignore',
                       suppress_warnings=True,
                       stepwise=True)

    # Fit the ARIMA model with the optimal p, d, and q values
    model.fit(df)

    # Forecast the test data using the ARIMA model
    forecast = model.predict(n_periods=24)

    # Create a dataframe to store the forecast results
    forecast_df = pd.DataFrame(forecast, columns=['price'])
    print(forecast_df.head())
    stacked_df = pd.concat([df, forecast_df])
    stacked_df['City'], stacked_df['State'] = city, state
    forecasting_ready_data.append(stacked_df)
    
    # Plot the forecast results with the price data
    plt.plot(df, label='Training Data')
    plt.plot(forecast_df, label='Forecast')
    plt.legend(loc='upper left')
    plt.title(str(city) + " "+ 'Rent Forecast')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.xticks(rotation=45)
    plt.show() 

forecasting_data = pd.concat(forecasting_ready_data)
```

```python
numeric_cols = df_merged.select_dtypes(include="number")
numeric_cols.drop(['zip', 'year_x', 'month'], axis=1, inplace=True)
```

```python
numeric_cols['City'] = df_merged['City']
numeric_cols = numeric_cols.groupby('City').mean()
```

```python
numeric_cols.head()
```

```python
plt.figure(figsize=(20, 15))
cmap = sns.diverging_palette(250, 10, as_cmap=True)
sns.heatmap(numeric_cols.corr(), vmax=.5, linewidths=.2, cmap=cmap)
plt.show()
```

```python
normalized_data = (numeric_cols.iloc[:,1:] - np.mean(numeric_cols.iloc[:,1:], axis=0)) / np.std(numeric_cols.iloc[:,1:], axis=0)
normalized_data.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_projection = tsne.fit_transform(normalized_data.iloc[:,:])

# Plot the t-SNE projection, labeling each point with the corresponding country name and hash
plt.figure(figsize=(10, 10))
for i in range(normalized_data.shape[0]):
    plt.scatter(tsne_projection[i, 0], tsne_projection[i, 1])
    plt.annotate(normalized_data.index.values[i], (tsne_projection[i, 0], tsne_projection[i, 1]))
plt.show()
```

```python
tsne_df = pd.DataFrame(tsne_projection, columns=["tsne1", "tsne2"])
tsne_df["City"] = normalized_data.index.values
```
